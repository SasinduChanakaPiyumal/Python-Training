```python
import random

j = 0
count = 0
step = 100
n_People = 100
l_position = []

while j < n_People:
    L = 10
    while True:
        if random.random() < 0.5:  # Probability threshold for selecting "Left" changed to 0.4
            L -= 1
        else:
            L += 1
        if L == 0:
            count += 1
            l_position.append(step)
            break
        step += 1
    j += 1

avg = sum(l_position) / n_People
print("Average steps to fall:", avg)
```
